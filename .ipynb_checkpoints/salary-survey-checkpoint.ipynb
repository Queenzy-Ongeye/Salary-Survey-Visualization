2. Dropped `['Job Context','Other currency', 'Income context']` columns because the number of rows with null values was higher than the other columns, filling the other textual columns with the `'Not provided'` values.

4. Filled the `'Salary Benefits'` column with a `(0)` value assuming that not all companies give annual salary benefits to their employees and `'State', 'City', 'Job Industry', 'Job title', 'Gender', 'Race', 'Education Level'` with unknown values.

```python
def fill_missing_values(df):
    column = ['State', 'City', 'Salary Benefits', 'Job Industry', 'Job title', 'Gender', 'Race', 'Education Level']
    for col in column:
        if col == 'Salary Benefits':
            df[col] = df[col].fillna(0).astype(int)
        else:
            df[col] = df[col].fillna('unkown')
    return df
```

3. New features
- Created an `'Hours per week'` column with the assumption of the average working hours being 40
- Created a `'Salary per Hour'` column with was gained after doing a division of the annual_salary by hours_per_week multiplied by the weeks_per_year
  ` ${cleaned_data['Salary per Hour'] = cleaned_data['Annual Salary']/ (cleaned_data["Hours per Week"] * weeks_per_year)}`
- Reason: The salary_per_hour feature provides a normalized measure of earnings based on time, allowing for fairer comparisons between employees who work different numbers of hours per week. It's especially useful for identifying discrepancies in pay or assessing the value of part-time versus full-time positions.
- Usage: This new feature can be used in various analyses, such as comparing hourly rates across industries or job titles.

```python
# Salary Per Hour
cleaned_data["Hours per Week"] = 40
weeks_per_year = 52
cleaned_data['Salary per Hour'] = cleaned_data['Annual Salary']/ (cleaned_data["Hours per Week"] * weeks_per_year)
cleaned_data['Salary per Hour']
```

3. Target encoding for Annual salary for each industry
- In this example, the 'Job Industry' categories ('Tech', 'Healthcare', 'Education') are encoded with the average salary for each industry.
This technique is particularly useful for handling categorical variables with high cardinality (many unique categories) and when you believe the mean target value is a good representation of the category's effect.

```python
# Target Encoding
mean_salary = cleaned_data.groupby('Job Industry')['Annual Salary'].mean().astype(int)
cleaned_data['Job Industry Encoded'] = cleaned_data['Job Industry'].map(mean_salary)

print("Target Encoded DataFrame:")
print(cleaned_data)
```

# Data Visualization
1. Plot one: it shows the top 15 cities with the highest level of employment.

3. Plot 2: It shows the education Level Distribution Across Top 10 Job Industries
4. Plot 3 : It shows the top industries with the highest annual salary rates

```python
print(plt.style.available)
```

```python
# Cities with high levels of employment
city_employment_counts = cleaned_data['City'].value_counts()

# Convert to a DataFrame for better readability and reset the index
city_employment_df = city_employment_counts.reset_index()
city_employment_df.columns = ['City', 'EmploymentCount']
# Sorting employment in descending order
city_employment_df = city_employment_df.sort_values(by='EmploymentCount', ascending=False)
top_n_cities = 15
top_cities_df = city_employment_df.head(top_n_cities)

# Plotting
plt.figure(figsize=(12, 8))
sns.barplot(x='EmploymentCount', y='City', data=top_cities_df, palette='viridis')
plt.title('Top Cities with the Highest Levels of Employment')
plt.xlabel('Number of Employees')
plt.ylabel('City')
plt.show()
```

```python
# Count the occurrences of each job industry to find the top 10
top_industries = cleaned_data['Job Industry'].value_counts().head(10)
# Filter the DataFrame to include only rows where 'Job Industry' is one of the top industries
top_industries_list = top_industries.index.tolist()
filtered_df = cleaned_data[cleaned_data['Job Industry'].isin(top_industries_list)]

# Group by industry and education level, and count the occurrences
education_by_industry = filtered_df.groupby(['Job Industry', 'Education Level']).size().unstack().fillna(0)

education_by_industry.T.plot(kind='bar', figsize=(14, 8), stacked=True)
plt.title('Education Level Distribution Across Top 10 Job Industries')
plt.xlabel('Education Level')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45)
plt.legend(title='Job Industry', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Top industries with the highest annual salary rates

mean_salary = cleaned_data.groupby('Job Industry')['Annual Salary'].mean()
cleaned_data['Job Industry Encoded'] = cleaned_data['Job Industry'].map(mean_salary)

# Use the encoded salary values for plotting
mean_salary_by_industry_encoded = cleaned_data.groupby('Job Industry Encoded')['Annual Salary'].mean().sort_values(ascending=False)

top_industries = 15
mean_salary_by_industry_encoded.head(top_industries).plot(
    kind='bar', figsize=(10, 8), color='skyblue'
)
plt.title('Top Industries with the Highest Average Encoded Salary')
plt.xlabel('Job Industry Encoded')
plt.ylabel('Average Encoded Salary')
plt.xticks(rotation=45)
plt.show()
```

```python
# Industries with high level employment
industry_employment_counts = cleaned_data['Job Industry'].value_counts()

# Convert to a DataFrame for better readability and reset the index
industry_employment_df = industry_employment_counts.reset_index()
industry_employment_df.columns = ['Job Industry', 'EmploymentCount']
# Sorting employment in descending order
city_employment_df = industry_employment_df.sort_values(by='EmploymentCount', ascending=False)
top_n_jobs = 15
top_jobs_df = industry_employment_df.head(top_n_jobs)

# Plotting
plt.figure(figsize=(12, 8))
sns.barplot(x='Job Industry', y='EmploymentCount', data=top_jobs_df, palette='viridis')
plt.title('Top Job Industries with the Highest Levels of Employment')
plt.xlabel('Indusry')
plt.xticks(rotation=45)
plt.ylabel('Number of Employees')
plt.show()
```
